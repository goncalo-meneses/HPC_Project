```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib import colormaps

plt.rcParams.update({'font.size': 14})
```

```python

iterations = []
errors = []

with open('./err_800.dat', 'r') as f:
    for line in f:
        if line.strip() and 'Number of iterations' in line:
            parts = line.split()
            n = int(parts[3])
            e = float(parts[5])
            iterations.append(n)
            errors.append(e)

iterations = np.array(iterations)
errors = np.array(errors)

colors = np.linspace(0, 1, len(iterations))

fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(iterations, errors, c=colors, cmap='viridis', s=10)
ax.set_yscale('log')

ax.set_xlabel('Iterations', fontsize=14)
ax.set_ylabel(r'$\delta_{\mathrm{global}}$', fontsize=14)
ax.set_xlim(0, max(iterations) + 50)
ax.set_ylim(1.1e-5, 1e1)

# Clean up spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('err_g800_w1p95.pdf')
plt.show()
```
