```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

gridsizes = np.array([200, 400, 800, 1000])
iterations = np.array([382, 1206, 3003, 3727])

def linear_model(n, gamma, delta):
    return gamma + delta * n

popt, _ = curve_fit(linear_model, gridsizes, iterations)
gamma, delta = popt

plt.figure(figsize=(8, 5))
plt.plot(gridsizes, iterations, 'o', markersize=8, label='Data')
plt.plot(gridsizes, linear_model(gridsizes, *popt), '-', label=fr'Fit: $\gamma$={gamma:.1f}, $\delta$={delta:.2f}')
plt.xlabel('Grid size $N$')
plt.ylabel('Iterations to convergence')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
